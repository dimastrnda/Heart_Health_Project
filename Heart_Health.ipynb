```python
# Pisah X dan y
X = df.drop(['HeartDisease'], axis=1)
y = df['HeartDisease']
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)
```

```python
# Daftar kolom
cat_cols = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
num_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']  # urutannya benar
```

```python
# Transformasi kategorikal
cat_transformer = OrdinalEncoder(
    categories=[['F', 'M'], ['ATA', 'NAP', 'TA', 'ASY'], ['Normal', 'LVH', 'ST'], ['N', 'Y'], ['Up', 'Down', 'Flat']],
    handle_unknown='use_encoded_value',
    unknown_value=-1)
```

```python
# ColumnTransformer
preprocessor = ColumnTransformer([
    ('cat', cat_transformer, cat_cols),
    ('scale', StandardScaler(), num_cols)
], remainder='passthrough')
```

```python
# Pipeline akhir
pipe = Pipeline([
    ('preprocess', preprocessor),
    ('classifier', SVC(probability=True))
])
```

```python
# Fit model
pipe.fit(X_train, y_train)
```

```python
# Simpan pipeline
with open('pipe.pkl', 'wb') as f:
    pickle.dump(pipe, f)

# Simpan daftar fitur (untuk app.py)
with open('features.pkl', 'wb') as f:
    pickle.dump(X.columns.tolist(), f)

# Simpan df (jika ingin menampilkan info pada Streamlit)
with open('df.pkl', 'wb') as f:
    pickle.dump(df, f)
```
